## Prueba 01
#### Probando web scraping de esta pagina: https://dockerlabs.es/

```python

url = "https://dockerlabs.es/"
respuesta = requests.get(url)
# Si es valido y tiene yn codigo de estado 200, exitoso
if respuesta.status_code ==200:
    soup = BeautifulSoup(respuesta.text, "html.parser") #en formato texto 
    # Hay que extraer la info que esta dentro de un div y se puede hacer click 
    maquinas = soup.find_all ("div", onclick=True)
    conteo_maquinas = 1
    autores = set() # es un conjunto
    # Recorremos la informacion disponible
    for maquina in maquinas:
        onclick_text = maquina['onclick']
        autor = onclick_text.split("'")[7]
        autores.add (autor)
        nombre_maquina = onclick_text.split("'")[1]
        conteo_maquinas +=1
        #print(nombre_maquina)
    print("Autores encontrados")
#     for autor in autores:
#         print(autor)
    #print(f"Numero total de maquinas {conteo_maquinas}")
    for maquina in maquinas:
        onclick_text = maquina['onclick']
        nombre = onclick_text.split("'")[1]
        dificultad = onclick_text.split("'")[3]# la' tambien se consedra como elemento
        autor = onclick_text.split("'")[7]
        print(f"{nombre} --> {dificultad} --> {autor}")
else:
    print(f"Hubo un error al hacer la peticion{respuesta.status_code}")
```

```python
url = "https://www.tiempo.com/cordoba.htm"
respuesta = requests.get(url)
soup = BeautifulSoup(respuesta.text, "html.parser")

dias = soup.find("ul", class_="grid-container-7 dias_w").find_all("li")
# Recorremos el clima extendido
for d in dias:
    # Dia (Hoy, Mañana, Domingo...)
    dia = d.find("span", class_="text-0").get_text(strip=True)

    # Fecha (14 Nov, 15 Nov…)
    fecha = d.find("span", class_="subtitle-m").get_text(strip=True)

    # Tipo de clima (alt del icono)
    clima = d.find("img", class_="simbW")["alt"]

    # Probabilidad de lluvia
    prob_lluvia = d.find("span", class_="probabilidad").get_text(strip=True) if d.find("span", class_="probabilidad") else "0%"

    # Precipitación (38 l/m2)
    precipitacion = d.find("span", class_="changeUnitR").get_text(strip=True) if d.find("span", class_="changeUnitR") else "0 l/m²"

    # Temperaturas
    temp_max = d.find("span", class_="max").get_text(strip=True)
    temp_min = d.find("span", class_="min").get_text(strip=True)

    # Viento
    viento_spans = d.find("span", class_="velocidad").find_all("span")
    viento_min = viento_spans[0].get_text(strip=True)
    viento_max = viento_spans[1].get_text(strip=True)
    viento_unidad = viento_spans[2].get_text(strip=True)

    print(f"{dia} {fecha}")
    print(f"  Clima: {clima}")
    print(f"  Lluvia: {prob_lluvia} – {precipitacion}")
    print(f"  Temp: {temp_max} / {temp_min}")
    print(f"  Viento: {viento_min}-{viento_max} {viento_unidad}")
    print("------------------------------------------------")
```

```python
url = "https://dolarhoy.com/"
respuesta = requests.get(url)
soup = BeautifulSoup(respuesta.text, "html.parser")

div_compra = soup.find("div", class_="compra")        # <- queda como tag
valor_compra  = div_compra.find("div", class_="val").get_text(strip=True)
varPorcentaje_compra  = div_compra.find("div", class_="var-porcentaje").get_text(strip=True)

print(f' Compra:{valor_compra}')
print(f' % Variacion :{varPorcentaje_compra}' )
div_venta = soup.find("div", class_="venta")        # <- queda como tag
valor_venta  = div_venta.find("div", class_="val").get_text(strip=True)
varPorcentaje_venta  = div_venta.find("div", class_="var-porcentaje").get_text(strip=True)
print(f' Venta :{valor_venta}')
print(f' % Variacion:{varPorcentaje_venta}')
```
